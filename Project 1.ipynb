Our most important data here will be the genre, because we can't recrerate a movie then the nex best hting would be to look 
at the genre which has most profits, is most popular, has the highest rating and which directors and writers are 
associated with the said genres

```python
# In order to see the correlations of the genres to the profits,popularity,rating,directors and writers
#we will first have to create a list containing all the genres without any repetition of genres

genres=[]
for i in data5['genres'].unique():
    genres.append(i)
genres
```

```python
#Now let us select the data we want to comapare to our genres and put it to a table

data6=data5.loc[:,('genres','Net_profit','averagerating','runtime_minutes','numvotes','directors','writers')]
```

```python
#This is a copy of the table above but we shall see why we it is special as we go on.
special_data6=((data6.sort_values(by='Net_profit',ascending=False)).reset_index()).drop('index',axis=1)
special_data6.head()
#The column we are dropping is the intial indexing of the data
```

```python
#Let us look at a preview of the data to ensure we have everything we need

data6.head(5)
```

Now the problem is getting the Net profit,averagerating,averageduration,numvotes,directors and writers associated with each genre considereing in our table above(data6 and special data6) they appear severally and not in an organized manner. To solve this Issue we shall take the following steps

```python
#Lets start getting the net profit of each genre
# we shall create a list that contains all the totals then later we shall append them to our list containing the genres
#we shall initialize the list here

data8=[]

#First let us make sure we iterate the table(data6) a minimum of the number of genres that exist
for x in range(0,len(genres)):
# b will be our variable that holds the total netprofit of each genre and resets once each genre has finished looking 
#for similar genres within the table
    b=0
    for y in range(0,len(data6)):
#since we are iterating through the table using the values in the genre list, then the resulting totals will be in the same
#order as the genres appear in the list
        if genres[x] == data6.loc[y,'genres']:            
            b=(data6.loc[y,'Net_profit']+b)
        else:
            continue
    data8.append(b)
data8                                                                                                                                                  
```

```python
len((data6.loc[(data6['genres']=='Action,Drama,Thriller')&(data6['Net_profit'])]).loc[:,'Net_profit'])
```

```python
# I manually calculated the total netprofit of all movies under the 'Action,Drama,Thriller' which contains 23 movies as 
#illustrated above, the total was 936850141 and to ensure our list working we shall see if our total is part of the list
#of the totals above

assert 936850141 in data8
```

```python
#Next we shall be calculating the average rating for each genre and for that we will need a function of getting averges of lists
def avg(lst):
    return sum(lst)/len(lst)
```

```python
#similiar to the loop we use to calculate the netprofit 
#we shall do the same be this time we are intrested in the ratings not the total so we will generate a list contaning the 
#ratings found in that genre then later find the average of each list within the list using the function above
data9=[]
for x in range(0,len(genres)):
    subset=[]
    for y in range(0,len(data6)):      
        if genres[x] == data6.loc[y,'genres']:            
            subset.append(data6.loc[y,'averagerating'])
        else:
            continue
    data9.append(subset)
data9
```

```python
#To get the average rating of each genre and storing the averages in a final list contaning the genres of each 
data10=[]
for i in data9:
    data10.append(round(avg(i),2))
data10
```

```python
#This loop gets the total number of votes of each genre and works in same way as the loops above
data11=[]

for x in range(0,len(genres)):
    b=0
    for y in range(0,len(data6)):      
        if genres[x] == data6.loc[y,'genres']:            
            b=(data6.loc[y,'numvotes']+b)
        else:
            continue
    data11.append(b)
data11    
```

```python
#We will run a similar code as we try to get the average duration of each movie
rrtdata=[]
for x in range(0,len(genres)):
    subset=[]
    for y in range(0,len(data6)):      
        if genres[x] == data6.loc[y,'genres']:            
            subset.append(data6.loc[y,'runtime_minutes'])
        else:
            continue
    rrtdata.append(subset)
rtdata=[]
for i in rrtdata:
    rtdata.append(round(avg(i),2))
rtdata
```

```python
#Remember our special table(special_data6), here instead of using the normal data6 table we have been using we shall use our 
#special table.
#The reason it is special is because the data in that table is orderd according to the Net profit in a descending order
#The reason is this, the Movies with the most profit will be first and so will the director associated with the movie, 
#therefore when the loop is 'looping' whichever the genre, the director assoisiated with the movie generating the most revenue
#of that genre will appear first in the list of directors assosiated with that genre
#And to avoid duplicates we shall loop thrice
direct=[]
for x in range(0,len(genres)):
    subset=[]
    for y in range(0,len(special_data6)):      
        if genres[x] == special_data6.loc[y,'genres']:
            if special_data6.loc[y,'directors'] in subset:
                continue
            else:
                subset.append(special_data6.loc[y,'directors'])
        else:
            continue
    direct.append(subset)
direct
```

```python
#This loop is similar to the one above only it is for writers specializing in those genres
writ=[]
for x in range(0,len(genres)):
    subset=[]
    for y in range(0,len(special_data6)):      
        if genres[x] == special_data6.loc[y,'genres']: 
            if special_data6.loc[y,'writers'] in subset:
                continue
            else:
                subset.append(special_data6.loc[y,'writers'])
        else:
            continue
    writ.append(subset)
writ
```

```python
#Now let's assign the values of the netprofit to the respective Genres
#we will do this through dictionaries
a={}
for i in range(0,len(genres)):
    a[genres[i]] = data8[i]
#Now, let's assign the values of the average rating to the respective Genres
b={}
for i in range(0,len(genres)):
    b[genres[i]]=data10[i]
#As discussed above, the higher number of votes implies that more people watched the movie and thus in this case we shall
#assign each genre to the total number of votes associated with it and consequently the popularity of the said genre.
c={}
for i in range(0,len(genres)):
    c[genres[i]]=data11[i]
#Now, let's assign the values of the average runtime to the respective Genres
d={}
for i in range(0,len(genres)):
    d[genres[i]]=rtdata[i]

#Now, let us look at the directors involved with this high performing genres

e={}
for i in range(0,len(genres)):
    e[genres[i]]=direct[i]
    
#Now, let's look at the writers working with the directors 

f={}
for i in range(0,len(genres)):
    f[genres[i]]=writ[i]

#Now let us create a new table with the above results assigned to their genres

data12= pd.DataFrame({'Genre':genres,
                      'Net_Profit':list(a.values()),
                      'Average_rating':list(b.values()),
                      'Num_votes':list(c.values()),
                      'Average_runtime':list(d.values()),
                      'Directors':list(e.values()),
                      'Writers':list(f.values())})
data12.head()
```

```python
#lets now sort the data

#we shall start with the 20 most profitable genres
data13= ((data12.sort_values(by='Net_Profit',ascending=False)).reset_index()).drop('index',axis=1)
data13.head(20)
```
